<b>What is a Sine Wave?</b> <br>
According to the wikipedia, "A sine wave or sinusoid is a mathematical curve that describes a smooth periodic oscillation. A sine wave is a continuous wave. It is named after the function sine, of which it is the graph. It occurs often in pure and applied mathematics, as well as physics, engineering, signal processing and many other fields."

```python
import pyaudio
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import IPython.display as ipd
```

# plotting a Simple Sine Wave

```python
# Plot a sine wave using time and amplitude obtained for the sine wave
sampleRate = 100
frequency = 1
audio_length = 1 #second
t = np.linspace(0,audio_length, sampleRate * audio_length)  #  Produces a given 'length' seconds Audio-File
y = np.sin(frequency * 2 * np.pi * t)  #  Has frequency of given 'frequency'

plt.plot(t, y)
plt.title('Sine wave')
plt.xlabel('Time')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
plt.show()
```

<b> AMPLITUDE </b>
<br>The amplitude of a wave refers to the maximum amount of displacement of a particle on the medium from its rest position. In a sense, the amplitude is the distance from rest to crest. Similarly, the amplitude can be measured from the rest position to the trough position. (A crest is a point on a wave with the maximum value upward displacement within a cycle. A trough is a minimum or lowest point in a cycle.)


```python
sampleRate = 44100
frequency = 10
audio_length = 1 #second
amplitude = 10

t = np.linspace(0, audio_length, sampleRate * audio_length)  #  Produces a given 'length' seconds Audio-File
y = amplitude*np.sin(frequency * 2 * np.pi * t)  #  Has frequency of given 'frequency'

plt.plot(t,y) # plot using pyplot library from matplotlib package
plt.title('Sine wave f='+str(frequency)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure

```

<b>FREQUENCY</b><br>
Frequency is a measure of the number of wave crests that pass a fixed point per second. At higher frequencies, more wave crests will pass the point each second; for this to happen, the wavelengths must be short. As frequency decreases, fewer waves will pass the fixed point per second, so wavelengths will be relatively longer.

```python
# generating specific 20htz audio of 1 second

sampleRate = 44100
frequency = 20
length = 1 #second

t = np.linspace(0, length, sampleRate * length)  
y = np.sin(frequency * 2 * np.pi * t)  

plt.plot(t,y) 
plt.title('Sine wave f='+str(frequency)+' Hz') 
plt.xlabel('Time (s)') 
plt.ylabel('Amplitude') 
plt.show()
```

A sample is a value or set of values at a point in time and/or space. A sampler is a subsystem or operation that extracts samples from a continuous signal. A theoretical ideal sampler produces samples equivalent to the instantaneous value of the continuous signal at the desired points.

<b> Using the Formula</b><br>

```python
def sine_wave(f,overSampRate,phase,nCyl):
    """
    f : frequency of sine wave in Hertz
    overSampRate : oversampling rate (integer)
    phase : desired phase shift in radians
    nCyl : number of cycles of sine wave to generate
    Returns:
    (t,g) : time base (t) and the signal g(t) as tuple
    """
    fs = overSampRate*f # sampling frequency
    t = np.arange(0,nCyl*1/f-1/fs,1/fs) # time base
    g = np.sin(2*np.pi*f*t+phase) # replace with cos if a cosine wave is desired
    return (t,g) # return time base and signal g(t) as tuple

"""
Simulate a sinusoidal signal with given sampling rate
"""
f = 2 #frequency = 10 Hz
overSampRate = 44000 #oversammpling rate
fs = f*overSampRate #sampling frequency
phase = 1/3*np.pi #phase shift in radians
nCyl = 5 # desired number of cycles of the sine wave

(t,x) = sine_wave(f,overSampRate,phase,nCyl) #function call

plt.plot(t,x) 
plt.title('Sine wave f='+str(f)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure
```
