```python
import pyaudio
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
import IPython.display as ipd
```

# Combining waves with same frequency


```python
sampleRate = 44100
audio_length = 1 #second
amplitude = 10

frequency_f = 10
time_t = np.linspace(0, audio_length, sampleRate * audio_length)  #  Produces a given 'length' seconds Audio-File

sine1 = amplitude*np.sin(frequency_f * 2 * np.pi * time_t)  #  Has frequency of given 'frequency'
sine2 = amplitude*np.sin(frequency_f * 2 * np.pi * time_t)  #  Has frequency of given 'frequency'

combined = sine1+sine2

plt.plot(time_t,sine1) # plot using pyplot library from matplotlib package
plt.title('Sine wave1 f='+str(frequency_f)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure

plt.plot(time_t,sine2) # plot using pyplot library from matplotlib package
plt.title('Sine wave2 f='+str(frequency_f)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure

plt.plot(time_t,combined) # plot using pyplot library from matplotlib package
plt.title('Sine wave1 + sine wave2') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure
```

Note: <br>
When we combined 2 similar sine waves, we can see that the amplitude increases. It got increased from 10 to 20 i.e 10+10

# Combining wave with different Frequencies

```python
sampleRate = 44100
audio_length = 1 #second
amplitude = 10

frequency_f = 10
frequency_f2 = 20

time_t = np.linspace(0, audio_length, sampleRate * audio_length)  #  Produces a given 'length' seconds Audio-File

sine1 = amplitude*np.sin(frequency_f * 2 * np.pi * time_t)  #  Has frequency of given 'frequency'
sine2 = amplitude*np.sin(frequency_f2 * 2 * np.pi * time_t)  #  Has frequency of given 'frequency'

combined = sine1+sine2

plt.plot(time_t,sine1) # plot using pyplot library from matplotlib package
plt.title('Sine wave1 f='+str(frequency_f)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure

plt.plot(time_t,sine2) # plot using pyplot library from matplotlib package
plt.title('Sine wave2 f='+str(frequency_f2)+' Hz') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure

plt.plot(time_t,combined) # plot using pyplot library from matplotlib package
plt.title('Sine wave1 + sine wave2') # plot title
plt.xlabel('Time (s)') # x-axis label
plt.ylabel('Amplitude') # y-axis label
plt.show() # display the figure
```
